```python
import re

def clean_weight_series(s: pd.Series) -> pd.Series:
    """
    Tar en Series med viktsträngar och returnerar en float Series med gram.
    - Konverterar 'kg' till gram (multiplicerar med 1000).
    - Tar bort 'g'/'gram' och parsar tal.
    - Negativa värden blir NaN.
    - Icke-parsbara värden blir NaN.
    """
    s = s.astype("string").str.strip().str.lower().fillna("")
    
    extracted = s.str.extract(r'(?P<val>[-+]?\d+[.,]?\d*)\s*(?P<unit>kg|kilo|kilogram|g|gram)?', expand=True)
    vals = extracted["val"].str.replace(",", ".").astype(float, errors="ignore")
    vals = pd.to_numeric(vals, errors="coerce")
    units = extracted["unit"].fillna("")

    grams = vals.copy()
    is_kg = units.str.contains(r'kg|kilo|kilogram', na=False)
    grams[is_kg] = grams[is_kg] * 1000.0
    grams = grams.where(grams >= 0, np.nan)

    return grams

df["vikt_gram_clean"] = clean_weight_series(df["vikt_gram"])

print(df[["vikt_gram", "vikt_gram_clean"]].value_counts(dropna=False).head(30))
print()
print(f"Datatyp: {df['vikt_gram_clean'].dtype}")
print(f"Antal (NaN): {df['vikt_gram_clean'].isna().sum()}")
```

```python
from transformers import pipeline
import pandas as pd
import torch

classifier = pipeline(
    "sentiment-analysis",
    model="nlptown/bert-base-multilingual-uncased-sentiment",
    device=1
)

test_result = classifier("Det här var toppen!")
print("Testresultat:", test_result)
```

```python
texts = df['recension_text'].fillna("").astype(str).tolist()

results = classifier(texts)

print("Analyserar recensioner...")
sentiments = []
for text in df['recension_text']:
    if pd.isna(text):
        sentiments.append({"sentiment": "neutral", "score": 0.0})
    else:
        result = classifier(str(text)[:512])[0]
        label = result['label']
        
        if label in ['5 stars', '4 stars']:
            sentiment = "positive"
        elif label == '3 stars':
            sentiment = "neutral"
        else:
            sentiment = "negative"
        
        sentiments.append({"sentiment": sentiment, "score": result['score']})


df['sentiment'] = [s['sentiment'] for s in sentiments]
df['sentiment_score'] = [s['score'] for s in sentiments]

print("\n=== RESULTAT ===")
print(df['sentiment'].value_counts())
print("\n=== EXEMPEL ===")
print(df[['produktnamn', 'sentiment', 'sentiment_score']].head(15))
```

```python
summarizer = pipeline(
    "text-generation",
    model="gpt2",
    device=-1
)

print("Skapar sammanfattningar...")
recension_clean = []

for text in df['recension_text']:
    if pd.isna(text) or len(str(text).split()) < 10:
        recension_clean.append(text)
    else:
        try:
            prompt = f"Sammanfatta: {str(text)[:512]}"
            summary = summarizer(prompt, max_new_tokens=50, min_length=10, do_sample=False)
            summary_text = summary[0]["generated_text"].replace(prompt, "").strip()
            recension_clean.append(summary_text if summary_text else text)
        except Exception:
            recension_clean.append(text)

df['recension_clean'] = recension_clean

print(df[['recension_clean', 'sentiment', 'sentiment_score']].head(15))
```

# Dokumentation: kolumner och transformationer efter datarengöring

| Kolumn | Datatyp | Transformation | Kommentar |

| wanted_leveransdatum_clean | datetime64[ns] | Tolkad som datum (blandade format, dayfirst=True). | Ogiltiga värden blir NaT. |
| faktiskt_leveransdatum_clean | datetime64[ns] | Tolkad som datum (blandade format, dayfirst=True). | Ogiltiga värden blir NaT. |
| recensionsdatum_clean | datetime64[ns] | Tolkad som datum. | Ogiltiga värden blir NaT. |
| produktnamn_clean | string | Trim + lowercase och normalisering via mapping. | Varianter samlas till gemensamt namn. |
| season_clean | string | Trim + lowercase och normalisering. | “tidig vår”, “early spring”, “q2” → “vår”. |

| vikt_gram_clean | float | Tolkad till gram; kg → gram. | Negativa/ogiltiga värden blir NaN. |

## Kvalitetsflaggor (sammanfattning)

| Flagg | Regel |
|---|---|
| flag_pris | Saknat/≤0 eller topp 1% i pris_clean. |
| flag_antal | Saknat/≤0 eller topp 1% i antal_clean. |
| flag_vikt | Saknat/≤0 eller topp 1% i vikt_gram_clean. |
| flag_datum | faktiskt_leveransdatum_clean tidigare än orderdatum_clean. |
| flag_total | Summa av flaggor per rad. |

## Sammanfattning
- Datum standardiserades och sattes till `datetime`.
- Textfält normaliserades (trim, lowercase och mapping).
- Numeriska fält rensades från text och konverterades till tal.
- Sentiment klassificerades med score för styrka.
- Kvalitetsflaggor skapades för orimliga/extrema värden.

## Extra justeringar
- **Produktnamn:** Målet var att minska antalet unika produktnamn. Jag tog bort onödiga tillägg och samlade varianter under ett gemensamt namn. Exempel: alla tomatfrön samlades under ett namn, och sort ses som ett valbart tillägg. Samma princip användes för t.ex. terrakottakruka där storlek inte längre separerar namnet.
- **Negativ vikt:** Negativa värden i `vikt_gram` sattes till NaN eftersom negativ vikt är orimlig och annars skulle snedvrida beräkningar (t.ex. medel och totaler).

## Val av modell (BERT vs LLM)
Jag valde BERT eftersom jag förstod uppgiften som att den ska vara kort, svensk textklassificering där en kompakt modell fungerar bra och är snabbare att köra lokalt. Det har gett mig stabila och reproducerbara resultat utan behov av API‑anrop, och att det kändes mer passande när det är många rader.

En rad per orderrad (unik `orderrad_id`), vilket bevarar relationen mellan order, produkt och kund på radnivå.

```python
import pandas as pd
import numpy as np
from pathlib import Path

OUTPUT_DIR = Path("data/processed")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

PRODUCT_MAPPING = {
    "terrakottakruka 25cm": "terrakottakruka",
    "terrakottakruka 15cm": "terrakottakruka",
    "tomatfrön 'biff'": "tomatfrön",
    "tommatfrön biff": "tomatfrön",
    "tomatfrön 'tiny tim'": "tomatfrön",
    "tommat tiny tim": "tomatfrön",
    "tomatplanta 'biff'": "tomatplanta",
    "tomatplanta 'tiny tim'": "tomatplanta",
    "örter mix (basilika, persilja, dill)": "örtfrön mix",
    "planteringsjord 25l": "planteringsjord",
    "planteringsjord 50l": "planteringsjord",
    "trädgårdspade": "trädgårdsspade",
    "flytande näring 1l": "flytande näring",
    "spade": "trädgårdsspade",
    "plastlåda balkong 60cm": "balkonglåda plast",
    "plastlåda 60cm": "balkonglåda plast",
    "gullranka ampel": "gullranka",
    "tomat tinytim": "tomatfrön",
    "terrakotta 25cm": "terrakottakruka",
    "sallads frön": "salladsfrön mix",
    "sallat mix": "salladsfrön mix",
    "biffftomat": "tomatfrön",
    "planteringsjord 50 liter": "planteringsjord",
    "sallatsfrön mix": "salladsfrön mix",
    "tomat-biff frön": "tomatfrön",
    "tomatffrön tiny tim": "tomatfrön",
    "petunia-mix": "petunia mix",
    "pettunia mix": "petunia mix",
    "röd pelargonner": "pelargon röd",
    "pelargoner röd": "pelargon röd",
    "örter mix": "örtfrön mix",
}

SEASON_MAPPING = {
    "tidig vår": "vår",
    "early spring": "vår",
    "q2": "vår"
}

ZONE_MAPPING = {
    "Zone 1": "Zon 1",
    "Zon1": "Zon 1",
    "Zone 5": "Zon 5",
    "Zon5": "Zon 5",
    "Zon2": "Zon 2",
}

EXPECTED_COLUMNS = [
    "orderdatum", "önskat_leveransdatum", "faktiskt_leveransdatum",
    "recensionsdatum", "produktnamn", "säsong", "antal", "pris",
    "vikt_gram", "leveranszon", "recension_text"
]

CLEAN_OUTPUT_COLUMNS = [
    "order_id", "orderrad_id", "produkt_id", "kund_id",
    "orderdatum_clean", "wanted_leveransdatum_clean",
    "faktiskt_leveransdatum_clean", "produktnamn_clean",
    "season_clean", "kategori", "antal_clean", "pris_clean",
    "vikt_gram_clean", "leveranszon_clean", "betyg",
    "sentiment", "sentiment_score", "recension_clean",
    "recensionsdatum_clean", "is_suspect"
]

def clean_weight_series(s: pd.Series) -> pd.Series:
    """Konverterar vikt till gram från olika format."""
    s = s.astype("string").str.strip().str.lower().fillna("")
    extracted = s.str.extract(
        r"(?P<val>[-+]?\d+[.,]?\d*)\s*(?P<unit>kg|kilo|kilogram|g|gram)?",
        expand=True,
    )
    vals = extracted["val"].str.replace(",", ".")
    vals = pd.to_numeric(vals, errors="coerce")
    units = extracted["unit"].fillna("")
    grams = vals.copy()
    is_kg = units.str.contains(r"kg|kilo|kilogram", na=False)
    grams[is_kg] = grams[is_kg] * 1000.0
    grams = grams.where(grams >= 0, np.nan)
    return grams


def apply_sentiment_analysis(df: pd.DataFrame) -> pd.DataFrame:
    """Applicerar sentiment-analys på recensioner."""
    sentiments = []
    
    if "classifier" not in globals():
        print("⚠ `classifier` saknas. Sätter neutral sentiment för alla rader.")
        sentiments = [{"sentiment": "neutral", "score": 0.0}] * len(df)
    else:
        for text in df["recension_text"]:
            if pd.isna(text):
                sentiments.append({"sentiment": "neutral", "score": 0.0})
            else:
                result = classifier(str(text)[:512])[0]
                label = result["label"]
                if label in ["5 stars", "4 stars"]:
                    sentiment = "positive"
                elif label == "3 stars":
                    sentiment = "neutral"
                else:
                    sentiment = "negative"
                sentiments.append({"sentiment": sentiment, "score": result["score"]})
    
    df["sentiment"] = [s["sentiment"] for s in sentiments]
    df["sentiment_score"] = [s["score"] for s in sentiments]
    return df


def summarize_reviews(df: pd.DataFrame) -> pd.DataFrame:
    """Sammanfattar recensioner med summarizer."""
    if "summarizer" not in globals():
        print("⚠ `summarizer` saknas. Behåller originaltext i recension_clean.")
        df["recension_clean"] = df["recension_text"]
    else:
        recension_clean = []
        for text in df["recension_text"]:
            if pd.isna(text) or len(str(text).split()) < 10:
                recension_clean.append(text)
            else:
                try:
                    prompt = f"Sammanfatta: {str(text)[:512]}"
                    summary = summarizer(prompt, max_new_tokens=50, min_length=10, do_sample=False)
                    summary_text = summary[0]["generated_text"].replace(prompt, "").strip()
                    recension_clean.append(summary_text if summary_text else text)
                except Exception:
                    recension_clean.append(text)
        df["recension_clean"] = recension_clean
    return df

def clean_null_values(df: pd.DataFrame) -> pd.DataFrame:
    """Fyller ut saknade värden med rimliga defaults"""
    text_cols = ["produktnamn", "säsong", "leveranszon"]
    for col in text_cols:
        if col in df.columns:
            df[col] = df[col].fillna("okänd")
    
    if "recension_text" in df.columns:
        df["recension_text"] = df["recension_text"].fillna("")
    
    return df

def create_mapping_file(df: pd.DataFrame, output_dir: Path) -> None:
    """Skapar mapping-fil för alla transformationer."""
    mapping_data = []
    
    mappings = [
        ("produktnamn", "produktnamn_clean"),
        ("säsong", "season_clean"),
        ("leveranszon", "leveranszon_clean")
    ]
    
    for orig_col, clean_col in mappings:
        if orig_col in df.columns and clean_col in df.columns:
            for orig, clean in zip(df[orig_col], df[clean_col]):
                if pd.notna(orig) and pd.notna(clean):
                    if str(orig).strip().lower() != str(clean).strip().lower():
                        mapping_data.append({
                            "column": orig_col,
                            "original": orig,
                            "cleaned": clean
                        })
    
    df_mapping = pd.DataFrame(mapping_data).drop_duplicates()
    df_mapping.to_csv(output_dir / "gronagarden_mapping.csv", index=False)
    print(f"✓ Sparad: gronagarden_mapping.csv ({len(df_mapping)} rader)")
    if len(df_mapping) > 0:
        print(df_mapping.head(10))
```

```python

def transform_orders(df: pd.DataFrame) -> pd.DataFrame:
    """
    Komplett rengöringspipeline för orderdata.
    Tar in rå data och returnerar helt städad data.
    """
    df_clean = df.copy()
    df_clean = clean_null_values(df_clean)
    
    for col in EXPECTED_COLUMNS:
        if col not in df_clean.columns:
            df_clean[col] = pd.NA
    
    date_columns = {
        "orderdatum": "orderdatum_clean",
        "önskat_leveransdatum": "wanted_leveransdatum_clean",
        "faktiskt_leveransdatum": "faktiskt_leveransdatum_clean",
        "recensionsdatum": "recensionsdatum_clean"
    }
    
    for orig, clean in date_columns.items():
        df_clean[clean] = pd.to_datetime(
            df_clean[orig],
            errors="coerce",
            format="mixed",
            dayfirst=True
        )
    
    df_clean["produktnamn_clean"] = (
        df_clean["produktnamn"]
        .astype("string")
        .str.strip()
        .str.lower()
        .replace(PRODUCT_MAPPING)
    )
    
    df_clean["season_clean"] = (
        df_clean["säsong"]
        .astype("string")
        .str.strip()
        .str.lower()
        .replace(SEASON_MAPPING)
    )
    
    df_clean["antal_clean"] = (
        df_clean["antal"]
        .astype("string")
        .str.replace("st", "", regex=False)
        .str.replace('""', "", regex=False)
        .str.strip()
    )
    df_clean["antal_clean"] = pd.to_numeric(df_clean["antal_clean"], errors="coerce")
    
    df_clean["pris_clean"] = (
        df_clean["pris"]
        .astype("string")
        .str.replace("SEK", "", regex=False)
        .str.replace("kr", "", regex=False)
        .str.replace(",", "", regex=False)
        .str.replace(":-", "", regex=False)
        .str.strip()
    )
    df_clean["pris_clean"] = pd.to_numeric(df_clean["pris_clean"], errors="coerce")
    
    df_clean["vikt_gram_clean"] = clean_weight_series(df_clean["vikt_gram"])
    
    df_clean["leveranszon_clean"] = (
        df_clean["leveranszon"]
        .astype("string")
        .str.strip()
        .str.title()
        .replace(ZONE_MAPPING)
    )
    
    df_clean = apply_sentiment_analysis(df_clean)
    df_clean = summarize_reviews(df_clean)
    
    df_clean["suspect_missing_orderdatum"] = df_clean["orderdatum_clean"].isna()
    df_clean["suspect_missing_pris"] = df_clean["pris_clean"].isna()
    df_clean["suspect_invalid_pris"] = df_clean["pris_clean"] <= 0
    df_clean["suspect_invalid_antal"] = df_clean["antal_clean"] <= 0
    df_clean["suspect_invalid_vikt"] = df_clean["vikt_gram_clean"] <= 0
    
    df_clean["is_suspect"] = (
        df_clean["suspect_missing_orderdatum"] |
        df_clean["suspect_missing_pris"] |
        df_clean["suspect_invalid_pris"] |
        df_clean["suspect_invalid_antal"] |
        df_clean["suspect_invalid_vikt"]
    )
    
    return df_clean


def process_and_save(df_raw: pd.DataFrame, output_dir: Path) -> pd.DataFrame:
    """
    Kör hela rengöringspipen och sparar resultaten.
    Returnerar det rena datasetet.
    """
    print(f"\n{'='*60}")
    print(f"STARTAR DATARENGÖRING: {len(df_raw)} rader")
    print(f"{'='*60}\n")
    
    df = df_raw.drop_duplicates()
    print(f"Efter borttagning av dubletter: {len(df)} rader\n")
    
    df_transformed = transform_orders(df)
    
    print("\n=== SUSPECT-SAMMANFATTNING ===")
    suspect_cols = [
        "suspect_missing_orderdatum",
        "suspect_missing_pris",
        "suspect_invalid_pris",
        "suspect_invalid_antal",
        "suspect_invalid_vikt",
        "is_suspect"
    ]
    print(df_transformed[suspect_cols].sum())
    
    output_cols = [c for c in CLEAN_OUTPUT_COLUMNS if c in df_transformed.columns]
    df_clean = df_transformed[~df_transformed["is_suspect"]][output_cols].copy()
    df_suspect = df_transformed[df_transformed["is_suspect"]].copy()
    
    print(f"\n=== DATASET-STORLEKAR ===")
    print(f"Ursprungligt dataset: {len(df)} rader")
    print(f"Rent dataset: {len(df_clean)} rader ({len(df_clean)/len(df)*100:.1f}%)")
    print(f"Misstänkta rader: {len(df_suspect)} rader ({len(df_suspect)/len(df)*100:.1f}%)\n")
    
    df_clean.to_csv(output_dir / "gronagarden_clean.csv", index=False)
    print(f"✓ Sparad: gronagarden_clean.csv ({len(df_clean)} rader)")
    
    df_suspect.to_csv(output_dir / "gronagarden_suspect.csv", index=False)
    print(f"✓ Sparad: gronagarden_suspect.csv ({len(df_suspect)} rader)\n")
    
    create_mapping_file(df_transformed, output_dir)
    
    print(f"\n{'='*60}")
    print("DATARENGÖRING KLAR!")
    print(f"{'='*60}\n")
    
    return df_clean

df_raw = pd.read_csv("gronagarden_data.csv")
df_clean = process_and_save(df_raw, OUTPUT_DIR)
```

```python
import pandas as pd
from pathlib import Path

def print_missing_values(df: pd.DataFrame, cols: list[str]) -> None:
    """Kontrollerar saknade värden i kritiska kolumner."""
    print("1. Kontrollerar saknade värden i kritiska kolumner...")
    for col in cols:
        if col not in df.columns:
            print(f"   ⚠ {col}: Kolumn saknas")
            continue
        missing = df[col].isna().sum()
        if missing == 0:
            print(f"   ✓ {col}: Inga saknade värden")
        else:
            print(f"   ⚠ {col}: {missing} saknade värden")
    print()


def print_dtype_check(df: pd.DataFrame, checks: dict[str, str]) -> None:
    """Kontrollerar att datatyper är korrekta."""
    print("2. Kontrollerar datatyper...")
    for col, expected in checks.items():
        if col not in df.columns:
            print(f"   ⚠ {col}: Kolumn saknas")
            continue
        
        dtype = df[col].dtype
        ok = False
        
        if expected == "datetime":
            ok = pd.api.types.is_datetime64_any_dtype(dtype)
        elif expected == "numeric":
            ok = pd.api.types.is_numeric_dtype(dtype)
        else:
            ok = str(dtype) == expected
        
        if ok:
            print(f"   ✓ {col}: Korrekt datatyp ({expected})")
        else:
            print(f"   ⚠ {col}: Fel datatyp ({dtype}, förväntat {expected})")
    print()


def print_value_checks(df: pd.DataFrame) -> None:
    """Kontrollerar att värden är rimliga."""
    print("3. Kontrollerar rimliga värden...")
    
    checks = [
        ("pris_clean", lambda x: (x < 0).any(), "Finns negativa värden", "Inga negativa värden"),
        ("antal_clean", lambda x: (x <= 0).any(), "Finns värden <= 0", "Alla värden > 0"),
        ("vikt_gram_clean", lambda x: (x < 0).any(), "Finns negativa värden", "Inga negativa värden")
    ]
    
    for col, check_func, fail_msg, success_msg in checks:
        if col not in df.columns:
            print(f"   ⚠ {col}: Kolumn saknas")
        elif check_func(df[col]):
            print(f"   ⚠ {col}: {fail_msg}")
        else:
            print(f"   ✓ {col}: {success_msg}")
    print()


def print_quality_checks(df: pd.DataFrame, label: str) -> None:
    """Kör alla kvalitetskontroller på ett dataset."""
    print(f"\n{'='*60}")
    print(f"KVALITETSKONTROLLER: {label}")
    print(f"{'='*60}\n")
    
    critical_cols = [
        "order_id", "orderrad_id", "orderdatum_clean",
        "antal_clean", "pris_clean"
    ]
    print_missing_values(df, critical_cols)
    
    dtype_checks = {
        "orderdatum_clean": "datetime",
        "pris_clean": "numeric",
        "antal_clean": "numeric",
        "vikt_gram_clean": "numeric",
        "sentiment_score": "numeric",
    }
    print_dtype_check(df, dtype_checks)
    
    print_value_checks(df)

date_cols = ["orderdatum_clean", "wanted_leveransdatum_clean", 
             "faktiskt_leveransdatum_clean", "recensionsdatum_clean"]

df_clean = pd.read_csv("data/processed/gronagarden_clean.csv", parse_dates=date_cols)
df_val_raw = pd.read_csv("gronagarden_validation.csv")

df_val = process_and_save(df_val_raw, OUTPUT_DIR)

print(f"\ndf_clean: {len(df_clean)} rader, {df_clean.shape[1]} kolumner")
print(f"df_val: {len(df_val)} rader, {df_val.shape[1]} kolumner")

print_quality_checks(df_clean, "Huvuddata (gronagarden_clean.csv)")
print_quality_checks(df_val, "Valideringsdata (gronagarden_validation.csv)")

print(f"\n{'='*60}")
print("VALIDERING SLUTFÖRD!")
print(f"{'='*60}\n")
```

```python
import sqlite3

def save_to_database(output_dir: Path, db_name: str = "gronagarden_pipeline.db") -> None:
    """Sparar alla CSV-filer till SQLite-databas i projektets root"""
    db_path = Path(db_name)  # Sparar i root istället för output_dir
    
    print(f"\n{'='*60}")
    print("SPARAR TILL SQLite-DATABAS")
    print(f"{'='*60}\n")
    
    conn = sqlite3.connect(db_path)
    
    files_to_save = [
        ("gronagarden_clean.csv", "orders_clean"),
        ("gronagarden_suspect.csv", "orders_suspect"),
        ("gronagarden_mapping.csv", "mapping")
    ]
    
    for csv_file, table_name in files_to_save:
        file_path = output_dir / csv_file
        if file_path.exists():
            df = pd.read_csv(file_path)
            df.to_sql(table_name, conn, if_exists="replace", index=False)
            print(f"✓ Tabell '{table_name}' sparad: {len(df)} rader")
        else:
            print(f"⚠ Fil saknas: {csv_file}")
    
    conn.close()
    print(f"\n✓ Databas sparad i root: {db_path.absolute()}")
    print(f"{'='*60}\n")

df_raw = pd.read_csv("gronagarden_data.csv")
df_clean = process_and_save(df_raw, OUTPUT_DIR)

save_to_database(OUTPUT_DIR)
```
